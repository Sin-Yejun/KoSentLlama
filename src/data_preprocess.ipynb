```python
import pandas as pd
import json

# CSV 파일 불러오기
csv_file_train = "/content/drive/MyDrive/Colab Notebooks/Llama3.1 8B/db/emotional_data/train/train.CSV"
df_train = pd.read_csv(csv_file_train, encoding='cp949')
csv_file_test = "/content/drive/MyDrive/Colab Notebooks/Llama3.1 8B/db/emotional_data/test/test.CSV"
df_test = pd.read_csv(csv_file_test, encoding='cp949')

train_output_file = "/content/drive/MyDrive/Colab Notebooks/Llama3.1 8B/db/emotional_data/train/train.jsonl"
test_output_file = "/content/drive/MyDrive/Colab Notebooks/Llama3.1 8B/db/emotional_data/test/test.jsonl"
```

```python
def transform_jsonl(df, output_file):
    # JSONL 변환을 위한 리스트
    jsonl_data = []

    for _, row in df.iterrows():
        messages = [{"role": "system", "content": "당신은 감정을 공감해주는 챗봇입니다."}]

        # 대화 쌍을 순서대로 추가
        for i in range(1, 4):  # 사람문장1~3, 시스템문장1~3
            # NaN을 안전하게 문자열로 처리
            user_text = str(row.get(f"사람문장{i}", "")).strip()
            assistant_text = str(row.get(f"시스템문장{i}", "")).strip()

            # 유효한 텍스트만 추가
            if user_text and user_text != "nan":  # 빈 문자열, 'nan' 제외
                messages.append({"role": "user", "content": user_text})
            if assistant_text and assistant_text != "nan":  # 빈 문자열, 'nan' 제외
                messages.append({"role": "assistant", "content": assistant_text})

        # 메시지가 시스템 메시지만 있는 경우 제외
        if len(messages) > 1:
            jsonl_data.append({"messages": messages})

    # JSONL 파일 저장
    jsonl_file = output_file
    with open(jsonl_file, "w", encoding="utf-8") as f:
        for entry in jsonl_data:
            json.dump(entry, f, ensure_ascii=False)
            f.write("\n")

    print(f"✅ 변환 완료! {jsonl_file} 파일을 확인하세요. (총 {len(jsonl_data)}개 대화)")

    # Output:
    # ✅ 변환 완료! db/emotional_data/train/train.jsonl 파일을 확인하세요. (총 51630 개 대화)
```

```python
transform_jsonl(df_train, train_output_file)
transform_jsonl(df_test, test_output_file)
```
